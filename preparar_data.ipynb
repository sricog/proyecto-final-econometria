# Preparación de variables

Las decisiones que se tomaron para la preparación de los datos son las siguientes
* Rellenar los valores vacíos de la variable "estu_tieneetnia" por el valor de "N". 
* Convertir variables categóricas a numéricas usando diferentes criterios: Todas las variables dicotómicas se transformaron en binarias 0-1. Las variables ordinales se les aplicó el tipo de dato categorico con su respectivo orden (**precaución con esto**). Por último, las variables nominales se convirtieron usando one-hot encoding.
* Convertir la variable de fecha de nacimiento por la edad del estudiante cuando presentó el examen y luego eliminar registros inconsistentes.
* Transformar la variable estu_repite en binaria como 0: no repite y 1: sí repite
* Juntar atributos jornada completa y jornada única como una sola

OJO, yo no eliminé los registros vacíos porque aún hay muchos datos faltantes y como aclaración, hay datos faltantes con el valor de -1, otros con NaN en algunas variables y en las variables dummies los valores vacíos son una fila de puros 0's

```python
data_convert = data_subset.copy()
data_convert.fillna({"estu_tieneetnia": "N"}, inplace=True)
```

```python
var_dicotomicas = ["estu_estudiante", "cole_area_ubicacion", "cole_naturaleza", "cole_bilingue",
                   "estu_discapacidad", "estu_genero", "estu_tieneetnia",
                   "fami_tienecomputador", "fami_tieneinternet", "fami_tieneserviciotv"]

var_categoricas = ["cole_caracter", "cole_jornada", "estu_dedicacioninternet", "estu_dedicacionlecturadiaria",
                  "estu_horassemanatrabaja", "fami_educacionmadre", "fami_educacionpadre", "fami_estratovivienda",
                  "fami_trabajolabormadre", "fami_trabajolaborpadre"]
```

```python
def age(born):
    today = pd.Timestamp("2024-08-11")
    return today.year - born.year - ((today.month, 
                                      today.day) < (born.month, 
                                                    born.day))

data_convert["estu_fechanacimiento"] = pd.to_datetime(data_convert["estu_fechanacimiento"], format="%d/%m/%Y", errors='coerce')
data_convert.dropna(subset=["estu_fechanacimiento"], inplace=True)
data_convert["estu_edad"] = data_convert["estu_fechanacimiento"].apply(age)
data_convert.drop(columns='estu_fechanacimiento', inplace=True)
data_convert = data_convert[(data_convert["estu_edad"] > 13) & (data_convert["estu_edad"] < 90)]
```

## Transformar variables dicotómicas a categóricas

```python
data_convert["estu_estudiante"] = pd.Categorical(data_convert["estu_estudiante"], categories=["ESTUDIANTE", "INDIVIDUAL"], ordered=True)
data_convert["estu_genero"] = pd.Categorical(data_convert["estu_genero"], categories=["M", "F"], ordered=True)
data_convert["estu_tieneetnia"] = pd.Categorical(data_convert["estu_tieneetnia"], categories=["N", "S"], ordered=True)

data_convert["cole_area_ubicacion"] = pd.Categorical(data_convert["cole_area_ubicacion"], categories=["URBANO", "RURAL"], ordered=True)
data_convert["cole_bilingue"] = pd.Categorical(data_convert["cole_bilingue"], categories=["N", "S"], ordered=True)
data_convert["cole_naturaleza"] = pd.Categorical(data_convert["cole_naturaleza"], categories=["OFICIAL", "NO OFICIAL"], ordered=True)
data_convert["estu_discapacidad"] = pd.Categorical(data_convert["estu_discapacidad"], categories=["N", "S"], ordered=True)

data_convert["fami_tienecomputador"] = pd.Categorical(data_convert["fami_tienecomputador"], categories=["No", "Si"], ordered=True)
data_convert["fami_tieneinternet"] = pd.Categorical(data_convert["fami_tieneinternet"], categories=["No", "Si"], ordered=True)
data_convert["fami_tieneserviciotv"] = pd.Categorical(data_convert["fami_tieneserviciotv"], categories=["No", "Si"], ordered=True)
```

```python
data_convert["grupo_edad"] = pd.cut(
                                data_convert["estu_edad"],
                                bins=[14, 20, 25, np.inf],
                                right=False,                     
                                labels=["14-19", "20-25", "25+"]
                            ).astype("category")

data_convert = data_convert.join(pd.get_dummies(data_convert["grupo_edad"], dtype=pd.Int8Dtype()))
```

```python
data_convert["estu_repite"] = data_convert["estu_repite"].where(data_convert["estu_repite"].isna() | 
                                                                (data_convert["estu_repite"] == 0), 1).astype(pd.Int8Dtype())
```

```python
# Variables ordinales
data_convert["estu_dedicacioninternet"] = pd.Categorical(data_convert['estu_dedicacioninternet'], categories=['No Navega Internet', '30 minutos o menos', 'Entre 30 y 60 minutos', 'Entre 1 y 3 horas', 'Más de 3 horas'], ordered=True)
data_convert["estu_dedicacionlecturadiaria"] = pd.Categorical(data_convert['estu_dedicacionlecturadiaria'], categories=['No leo por entretenimiento', '30 minutos o menos', 'Entre 30 y 60 minutos', 'Entre 1 y 2 horas', 'Más de 2 horas'], ordered=True)
data_convert["estu_horassemanatrabaja"] = pd.Categorical(data_convert['estu_horassemanatrabaja'], categories=['0', 'Menos de 10 horas', 'Entre 11 y 20 horas', 'Entre 21 y 30 horas', 'Más de 30 horas'], ordered=True)
data_convert["fami_estratovivienda"] = pd.Categorical(data_convert['fami_estratovivienda'],
                                                     categories=['Sin Estrato', 'Estrato 1', 'Estrato 2', 'Estrato 3', 'Estrato 4', 'Estrato 5', 'Estrato 6'], ordered=True)

# Variables nominales
data_convert = data_convert.join(pd.get_dummies(data_convert["cole_caracter"], dtype=pd.Int8Dtype()))

data_convert["cole_jornada"] = data_convert["cole_jornada"].replace('COMPLETA', 'UNICA')
data_convert = data_convert.join(pd.get_dummies(data_convert["cole_jornada"], dtype=pd.Int8Dtype()))

# Educación padres
data_convert = data_convert.join(pd.get_dummies(data_convert["fami_educacionmadre"], dtype=pd.Int8Dtype()).rename(columns=lambda c: c.replace(c,"madre_edu " + c)))
data_convert = data_convert.join(pd.get_dummies(data_convert["fami_educacionpadre"], dtype=pd.Int8Dtype()).rename(columns=lambda c: c.replace(c,"padre_edu " + c)))

orden = ['No sabe', 'Ninguno', 'Primaria incompleta', 'Primaria completa', 
                  'Secundaria (Bachillerato) incompleta', 'Secundaria (Bachillerato) completa',
                  'Técnica o tecnológica incompleta', 'Técnica o tecnológica completa',
                  'Educación profesional incompleta', 'Educación profesional completa',
                 'Postgrado']
data_convert["fami_educacionmadre"] = pd.Categorical(data_convert['fami_educacionmadre'], categories=orden, ordered=True)
data_convert["fami_educacionpadre"] = pd.Categorical(data_convert['fami_educacionpadre'], categories=orden, ordered=True)


# Trabajo padres
data_convert = data_convert.join(pd.get_dummies(data_convert["fami_trabajolabormadre"], dtype=pd.Int8Dtype()).rename(columns=lambda c: c.replace(c,"madre_trabajo " + c)))
data_convert = data_convert.join(pd.get_dummies(data_convert["fami_trabajolaborpadre"], dtype=pd.Int8Dtype()).rename(columns=lambda c: c.replace(c,"padre_trabajo " + c)))
```

```python
data_convert["estu_nse_individual"] = data_convert["estu_nse_individual"].astype("Int8")
data_convert["estu_edad"] = data_convert["estu_edad"].astype("int8")
data_convert["punt_ingles"] = data_convert["punt_ingles"].astype("Int16")
data_convert["punt_lectura_critica"] = data_convert["punt_lectura_critica"].astype("Int16")
data_convert["punt_matematicas"] = data_convert["punt_matematicas"].astype("Int16")
data_convert["punt_sociales_ciudadanas"] = data_convert["punt_sociales_ciudadanas"].astype("Int16")
data_convert["punt_c_naturales"] = data_convert["punt_c_naturales"].astype("Int16")
data_convert["punt_global"] = data_convert["punt_global"].astype("Int16")
```

```python
# 1) Cargar shapes de departamentos de Colombia
gdf = gpd.read_file("Colombia.geo.json")
```

```python
recode_geo = {"BOGOTÁ, D.C.": "BOGOTA",
              "ARCHIPIÉLAGO DE SAN ANDRÉS, PROVIDENCIA Y SANTA CATALINA": "SAN ANDRES",
              "NORTE DE SANTANDER": "NORTE SANTANDER",
              "VALLE DEL CAUCA": "VALLE"}
gdf["dpto_cnmbr"] = gdf["dpto_cnmbr"].replace(recode_geo)
```

```python
def norm_series(s: pd.Series) -> pd.Series:
    return (s.astype(str)
              .str.upper()
              .map(lambda x: unidecode(x))
              .str.replace(r"[^A-Z\s\.]", "", regex=True)
              .str.replace(r"\s+", " ", regex=True)
              .str.strip())

gdf["dep_norm"] = norm_series(gdf["dpto_cnmbr"])
```

```python
data_subset["dep_norm"] = norm_series(data_subset["cole_depto_ubicacion"])
```

```python
data_subset["dep_norm"].unique()
```

```python
# 5) Frecuencias por departamento
freq = (data_subset.groupby("dep_norm")
          .size()
          .rename("n")
          .reset_index())

# 6) Unir frecuencias al GeoDataFrame
gdf_plot = gdf.merge(freq, on="dep_norm", how="left")
gdf_plot["n"] = gdf_plot["n"].fillna(0).astype(int)
```

```python
# 7) Graficar coroplético
fig, ax = plt.subplots(figsize=(9, 10))
gdf_plot.plot(
    column="n",
    cmap="OrRd",
    linewidth=0.6,
    edgecolor="black",
    legend=True,
    # scheme="Quantiles",  # alternativas: "NaturalBreaks", "EqualInterval"
    k=5,
    missing_kwds={"color": "lightgrey", "hatch": "///", "label": "Sin datos"},
    ax=ax,
)
ax.set_title("Participación de estudiantes por departamento", pad=10, fontsize=12)

# 8) (Opcional) Etiquetas con conteo en el centro geométrico
# pts = gdf_plot.representative_point()
# for x, y, val in zip(pts.x, pts.y, gdf_plot["n"]):
#     if val > 0:
#         ax.text(x, y, f"{val}", ha="center", va="center", fontsize=7)

plt.tight_layout()
plt.show()
```
