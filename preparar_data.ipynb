# Preparación de variables

Convertir variables categóricas a numéricas usando diferentes criterios

cole_cod_depto_ubicacion      (no tenemos el diccionario de variables)
estu_repite 		      (tratar todos los valores diferentes de 0)     
fami_educacionmadre	      (NO está ordenado)
fami_educacionpadre	      (NO está ordenado)
fami_trabajolabormadre	      (NO está ordenado)
fami_trabajolaborpadre	      (NO está ordenado)

```python
data_convert = data_subset.copy()

data_convert["estu_estudiante"] = pd.Categorical(data_convert["estu_estudiante"], categories=["INDIVIDUAL", "ESTUDIANTE"], ordered=True).codes

data_convert["cole_area_ubicacion"] = pd.Categorical(data_convert["cole_area_ubicacion"], categories=["RURAL", "URBANO"], ordered=True).codes
data_convert["cole_bilingue"] = pd.Categorical(data_convert["cole_bilingue"], categories=["N", "S"], ordered=True).codes
data_convert["cole_caracter"] = pd.Categorical(data_convert['cole_caracter'], categories=['ACADÉMICO', 'TÉCNICO', 'TÉCNICO/ACADÉMICO', 'NO APLICA']).codes
data_convert["cole_jornada"] = pd.Categorical(data_convert['cole_jornada'], categories=['UNICA', 'MAÑANA', 'COMPLETA', 'NOCHE', 'TARDE', 'SABATINA']).codes
data_convert["cole_naturaleza"] = pd.Categorical(data_convert["cole_naturaleza"], categories=["NO OFICIAL", "OFICIAL"], ordered=True).codes

data_convert["estu_dedicacioninternet"] = pd.Categorical(data_convert['estu_dedicacioninternet'], categories=['No Navega Internet', '30 minutos o menos', 'Entre 30 y 60 minutos', 'Entre 1 y 3 horas', 'Más de 3 horas'], ordered=True).codes
data_convert["estu_dedicacionlecturadiaria"] = pd.Categorical(data_convert['estu_dedicacionlecturadiaria'], categories=['No leo por entretenimiento', '30 minutos o menos', 'Entre 30 y 60 minutos', 'Entre 1 y 2 horas', 'Más de 2 horas'], ordered=True).codes
data_convert["estu_discapacidad"] = pd.Categorical(data_convert["estu_discapacidad"], categories=["N", "S"], ordered=True).codes
data_convert["estu_genero"] = pd.Categorical(data_convert["estu_genero"], categories=["M", "F"], ordered=True).codes
data_convert["estu_horassemanatrabaja"] = pd.Categorical(data_convert['estu_horassemanatrabaja'], categories=['0', 'Menos de 10 horas', 'Entre 11 y 20 horas', 'Entre 21 y 30 horas', 'Más de 30 horas'], ordered=True).codes
data_convert["estu_tieneetnia"] = pd.Categorical(data_convert["estu_tieneetnia"], categories=["N", "S"], ordered=True).codes

data_convert["fami_educacionmadre"] = pd.Categorical(data_convert['fami_educacionmadre']).codes # no está ordenado
data_convert["fami_educacionpadre"] = pd.Categorical(data_convert['fami_educacionpadre']).codes # no está ordenado
data_convert["fami_estratovivienda"] = pd.Categorical(data_convert['fami_estratovivienda'],
                                                      categories=['Sin Estrato', 'Estrato 1', 'Estrato 2', 'Estrato 3', 'Estrato 4', 'Estrato 5', 'Estrato 6'], ordered=True).codes
data_convert["fami_tienecomputador"] = pd.Categorical(data_convert["fami_tienecomputador"], categories=["N", "S"], ordered=True).codes
data_convert["fami_tieneinternet"] = pd.Categorical(data_convert["fami_tieneinternet"], categories=["N", "S"], ordered=True).codes
data_convert["fami_tieneserviciotv"] = pd.Categorical(data_convert["fami_tieneserviciotv"], categories=["N", "S"], ordered=True).codes
data_convert["fami_trabajolabormadre"] = pd.Categorical(data_convert['fami_trabajolabormadre']).codes # no está ordenado
data_convert["fami_trabajolaborpadre"] = pd.Categorical(data_convert['fami_trabajolaborpadre']).codes # no está ordenado
```

```python
data_convert[["cole_cod_depto_ubicacion", "estu_nse_individual", "punt_lectura_critica", "punt_matematicas", "punt_c_naturales", "punt_ingles", "punt_sociales_ciudadanas"]]= data_convert[["cole_cod_depto_ubicacion", "estu_nse_individual", "punt_lectura_critica", "punt_matematicas", "punt_c_naturales", "punt_ingles", "punt_sociales_ciudadanas"]].fillna(-1).astype('int8')
data_convert["punt_global"] = data_convert["punt_global"].fillna(-1).astype('int16')
```

Hay algunos errores en las fechas de nacimiento tales como 30/11/0002 los cuales impiden que la variable sea convertida a formato fecha.
Por lo tanto, se procede a eliminar estos resgistros, manteniendo solo las fechas válidas. Después, se crea una variable adicional para denotar la edad

```python
data_convert["estu_fechanacimiento"] = pd.to_datetime(data_convert["estu_fechanacimiento"], format="%d/%m/%Y", errors='coerce')
data_convert["estu_edad"] = (pd.to_datetime("2024-08-11") - data_convert["estu_fechanacimiento"]).dt.days // 365
data_convert["estu_edad"] = data_convert["estu_edad"].fillna(-1).astype('int8')
```
